```python
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
```

### Generate some data:

```python
from sklearn import datasets

N       = 500
mX, vY  = datasets.make_circles(n_samples=N, noise=.01)

mX[0,:] = [0,     0.1]
mX[1,:] = [-0.1, -0.1]
mX[2,:] = [0.1,  -0.1]
vY[:3]  = 0

vIdx0 = vY == 0
vIdx1 = vY == 1

plt.figure (figsize=(4, 4))
plt.scatter(mX[vIdx0,0], mX[vIdx0,1], s=50, color = 'r', edgecolor='k')
plt.scatter(mX[vIdx1,0], mX[vIdx1,1], s=50, color = 'b', edgecolor='k')
plt.axis   ('equal')
plt.show   ()
```

### Plot tree function:

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

#-- 2D grid:
v0       = np.linspace(mX[:,0].min(), mX[:,0].max(), 150)
v1       = np.linspace(mX[:,1].min(), mX[:,1].max(), 150)
XX0, XX1 = np.meshgrid(v0, v1)
XX       = np.c_[XX0.ravel(), XX1.ravel()]

def PlotTree(K):
    oTreeClassifier = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=K, random_state=0)
    oTreeClassifier.fit(mX, vY)

    Z = oTreeClassifier.predict(XX)
    Z = Z.reshape(XX0.shape)

    _, ax = plt.subplots(1, 2, figsize=(16, 8))
    ax[0].contourf(XX0, XX1, Z, colors=['red', 'blue'], alpha=0.3, levels=[-1.5, 0, 1.5])
    ax[0].scatter(mX[vIdx0,0], mX[vIdx0,1], s=75, color = 'r', edgecolor='k')
    ax[0].scatter(mX[vIdx1,0], mX[vIdx1,1], s=75, color = 'b', edgecolor='k')

    plot_tree(oTreeClassifier, filled=True, ax=ax[1], rounded=True)
    ax[1].set_title(f'Max splits = {K-1}')
    
```

```python
from ipywidgets import interact, IntSlider, Layout

kSlider = IntSlider(min=2, max=20, step=1, value=2, layout=Layout(width='80%'))
interact(PlotTree, K=kSlider)
plt.show()
```

### Generate some train and test data (from the same distribution):

```python
from sklearn.model_selection import train_test_split

N1  = 500
N2  = 500
mX1 = np.random.randn(N1, 2) + 1
mX2 = np.random.randn(N2, 2) - 1
vY1 = np.ones(N1)
vY2 = np.zeros(N2)
mX  = np.r_[mX1, mX2]
vY  = np.r_[vY1, vY2]

vIdx0 = vY == 0
vIdx1 = vY == 1

plt.figure (figsize=(4, 4))
plt.scatter(mX[vIdx0,0], mX[vIdx0,1], s=50, color = 'r', edgecolor='k')
plt.scatter(mX[vIdx1,0], mX[vIdx1,1], s=50, color = 'b', edgecolor='k')
plt.title  ('Train + test sets')
plt.axis   ('equal')
plt.show   ()

mTrainX, mTestX, vTrainY, vTestY = train_test_split(mX, vY, test_size=0.5)
```

### Train decision trees with various depths

```python
lTrees    = []
maxSplits = 75
for kk in range(1, maxSplits+1):
    oTree = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=kk+1, random_state=0)
    oTree.fit(mTrainX, vTrainY)
    lTrees.append(oTree)
```

### Plot function:

```python
vTrainRes = np.array([oTree.score(mTrainX, vTrainY) for oTree in lTrees])
vTestRes  = np.array([oTree.score(mTestX,  vTestY)  for oTree in lTrees])

v0       = np.linspace(mX[:,0].min(), mX[:,0].max(), 150)
v1       = np.linspace(mX[:,1].min(), mX[:,1].max(), 150)
XX0, XX1 = np.meshgrid(v0, v1)
XX       = np.c_[XX0.ravel(), XX1.ravel()]

def PlotTree(K):
    Z = lTrees[K-1].predict(XX)
    Z = Z.reshape(XX0.shape)

    _, ax = plt.subplots(1, 2, figsize=(16, 8))
    ax[0].contourf(XX0, XX1, Z, colors=['red', 'blue'], alpha=0.3, levels=[-1.5, 0, 1.5])
    ax[0].scatter(mX[vIdx0,0], mX[vIdx0,1], s=75, color = 'r', edgecolor='k')
    ax[0].scatter(mX[vIdx1,0], mX[vIdx1,1], s=75, color = 'b', edgecolor='k')

    ax[1].plot     (range(1, K+1), 1 - vTrainRes[:K], color='m', lw=2, marker='.', markersize=20, label='Train Error')
    ax[1].plot     (range(1, K+1), 1 - vTestRes [:K], color='k', lw=2, marker='.', markersize=20, label='Test Error' )
    ax[1].set_title(f'Max splits = {K}')
    ax[1].legend   ()
```

#### Increasing the depth of a decision tree will cause to overfit

```python
kSlider = IntSlider(min=1, max=maxSplits, step=1, value=1, layout=Layout(width='80%'))
interact(PlotTree, K=kSlider)
plt.show()
```
