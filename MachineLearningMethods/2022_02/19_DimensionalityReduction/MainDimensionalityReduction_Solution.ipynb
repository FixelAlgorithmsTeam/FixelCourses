```python
import numpy             as np
import pandas            as pd
import seaborn           as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine, load_breast_cancer, load_digits
from sklearn.manifold import TSNE, SpectralEmbedding, Isomap
from umap             import UMAP
```

### Exercies:
* Apply Isomap, Laplacian Eigenmaps (SpectralEmbedding), t-SNE and UMAP to:
    * The breast cancer dataset
    * The wine dataset
    * The digit dataset
    
* Try with and without features normalization.
* You may play with the algorithms' hyper-parameters.

```python
from sklearn.preprocessing import StandardScaler

lData      = [load_breast_cancer, load_wine, load_digits]
lDataStr   = ['Breast cancer',    'Wine',    'Digits'] 

lMethod    = [Isomap(),  SpectralEmbedding(),  TSNE(), UMAP()]
lMethodStr = ['Isomap', 'LaplacianEigenmaps', 'TSNE', 'UMAP']
lC         = ['r', 'g', 'b']

fig, ax = plt.subplots(3, 4, figsize=(12, 8))

for ii, Data in enumerate(lData):
    mX, vY = Data(return_X_y=True)
    
    if lDataStr[ii] != 'Digits':
        mX = StandardScaler().fit_transform(mX)
    
    for jj, oMethod in enumerate(lMethod):
        mZ = oMethod.fit_transform(mX)
        
        ax[ii,jj].scatter  (*mZ.T, s=25, c=vY, edgecolor='k', cmap='tab10', vmin=-1/2, vmax=9.5)                         
        ax[ii,jj].set_title(f'{lDataStr[ii]} - {lMethodStr[jj]}', color=lC[ii])
    
plt.tight_layout()
plt.show        ()
```
