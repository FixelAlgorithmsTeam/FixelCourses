```python
import numpy             as np
import matplotlib.pyplot as plt
import matplotlib.cm
```

```python
mX = np.load('clusterable_data.npy')

plt.figure (figsize=(7, 6))
plt.scatter(mX[:,0], mX[:,1], s=25, color=[0,1,0], edgecolor='k')
plt.xlabel ('$x_1$')
plt.ylabel ('$x_2$')
plt.axis   ('equal')
plt.grid   ()
plt.show   ()
```

```python
from hdbscan import HDBSCAN

oHDBSCAN = HDBSCAN(min_cluster_size=20, min_samples=5)
vC       = oHDBSCAN.fit_predict(mX)
np.unique(vC)
```

```python
Nc = len(np.unique(vC))

plt.figure(figsize=(7, 6))
plt.scatter(mX[:,0], mX[:,1], s=25, c=vC, edgecolor='k', cmap=matplotlib.cm.get_cmap('gist_rainbow', Nc))
plt.xlabel ('$x_1$')
plt.ylabel ('$x_2$')
plt.axis   ('equal')
plt.show   ()
```
