````python
# General Auxiliary Functions

def PlotSVR( vX: NDArray, vY: NDArray, vP: Optional[NDArray] = None, paramC: float = 1.0, ε: float = 0.1,
             numGridPts: int = 1001, hA: Optional[plt.Axes] = None, figSize: Tuple[int, int] = FIG_SIZE_DEF, 
             markerSize: int = MARKER_SIZE_DEF, lineWidth: int = LINE_WIDTH_DEF, axisTitle: str = None ) -> None:

    if hA is None:
        hF, hA = plt.subplots(figsize = figSize)
    else:
        hF = hasattr.get_figure()

    numSamples = len(vY)

    oMdl = SVR(kernel = 'linear', C = paramC, epsilon = ε).fit(vX[:, None], vY)
    
    # R2 Score
    vYHat = oMdl.predict(vX[:, None])
    valR2 = oMdl.score(vX[:, None], vY)
    vIdxSV = oMdl.support_
    
    # Plot
    vXX = np.linspace(np.around(np.min(vX), decimals = 1) - 0.1, np.around(np.max(vX), decimals = 1) + 0.1, numGridPts)
    vYY = oMdl.predict(vXX[:, None])

    hA.plot(vX, vY, '.r', ms = markerSize, label = r'$y_i$')
    hA.plot(vXX, vYY, 'b',  lw = lineWidth, label = r'$\hat{f}(x)$')
    hA.fill_between(vXX, vYY - ε, vYY + ε, color = 'b', alpha = 0.10, label = r'$\epsilon$ Tube')
    hA.plot(vXX, vYY - ε, '--b', lw = lineWidth, alpha = 0.60)
    hA.plot(vXX, vYY + ε, '--b', lw = lineWidth, alpha = 0.60)
    hA.plot(vX[vIdxSV], vY[vIdxSV], 'o', ms = 1.5 * markerSize, mfc = 'none', mec = 'k', mew = 1.5, label = 'Support Vectors')
    hA.set_title (f'R2 = {valR2:0.3f}')
    hA.set_xlabel('$x$')
    hA.set_xlim(left = vXX[0], right = vXX[-1])
    hA.set_ylim(bottom = np.floor(np.min(vY)), top = np.ceil(np.max(vY)))
    hA.grid()
    hA.legend()
````

## Support Vector Regression

The objective function of the Support Vector Regression (SVR) Model:

$$
\arg \min_{\boldsymbol{w}, b} \frac{1}{2} {\left\| \boldsymbol{w} \right\|}^{2} + C \sum_{i} \max \left\{ 0, \left| {y}_{i} - \left( \boldsymbol{w}^{T} \boldsymbol{x}_{i} - b \right) \right| - \epsilon \right\}
$$

 * $\boldsymbol{w}$ - The linear parameters of the model.
 * $b$ - The intercept parameter of the model.

The term $\sum_{i} \max \left\{ 0, \left| {y}_{i} - \left( \boldsymbol{w}^{T} \boldsymbol{x}_{i} - b \right) \right| - \epsilon \right\}$ is called $\epsilon$ _Insensitive Loss Function_ which makes the model outlier robust to some level.  

As the Support Vector Classifier (SVC) the SVR model supports utilization of the [_Kernel Trick_](https://en.wikipedia.org/wiki/Kernel_method).

* <font color='brown'>(**#**)</font> See [StackExchnage Statistics (Cross Validated) - Support Vector Regression vs. Linear Regression](https://stats.stackexchange.com/questions/633091), [MATLAB Help Center - Understanding Support Vector Machine Regression](https://www.mathworks.com/help/stats/understanding-support-vector-machine-regression.html).
* <font color='brown'>(**#**)</font> The robustness to _outliers_ is the result of the loss function being based on ${L}_{1}$ based loss.
* <font color='brown'>(**#**)</font> The [Ranking SVM](https://en.wikipedia.org/wiki/Ranking_SVM) is a variant of Support Vector Machine for Ranking.

````python
# Parameters

# Data Generation
tuXSupport  = (0, 5)
numSamples  = 75
noiseStd    = 0.45
numOutliers = 10
outlierStd  = 1.75

# Model
polyDeg = 2
λ       = 0.1

# Data Visualization
gridNoiseStd = 0.05
numGridPts = 250
````

## Generate / Load Data

The data is modeled as a real world phenomenon:

> A cheap sensor measures voltage $x$, and we want to predict physical temperature $y$ which a linear function of the voltage.  
> Most readings are accurate which an additive bounded noise and a few readings are corrupted.

````python
# Generate Data

vX = np.random.uniform(tuXSupport[0], tuXSupport[1], numSamples)

vIdxOutliers      = np.random.choice(numSamples, numOutliers, replace = False)
vY                = 1.25 * vX + 1 + np.random.normal(0, noiseStd, numSamples)
vY[vIdxOutliers] += np.random.normal(0, outlierStd, numOutliers)

print(f'The features data shape: {vX.shape}')
print(f'The labels data shape: {vY.shape}')
````

### Plot Regressor for Various Regularization (λ) Values

Let's see the effect of the strength of the regularization on the data.

````python
hPolyFitLasso = lambda paramC, ε: PlotSVR(vX, vY, paramC = paramC, ε = ε)
paramCSlider  = FloatSlider(min = 0.05, max = 5, step = 0.05, value = 1.0, readout_format = '0.3f', layout = Layout(width = '30%'))
epsilonSlider = FloatSlider(min = 0.01, max = 1, step = 0.01, value = 0.1, readout_format = '0.3f', layout = Layout(width = '30%'))
interact(hPolyFitLasso, paramC = paramCSlider, ε = epsilonSlider, continuous_update = False);
````
