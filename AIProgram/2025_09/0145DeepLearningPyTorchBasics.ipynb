## PyTorch

In our days _PyTorch_ is considered to be the _Go To_ Deep Learning framework.

![Papers with Code: PyTorch vs. TensorFlow](https://i.imgur.com/BybdtbK.png)

Source [AssemblyAI - PyTorch vs TensorFlow in 2023](https://www.assemblyai.com/blog/pytorch-vs-tensorflow-in-2023/).

Modern DL framework is composed of the following components:

 - **Data Structure**  
   The container of the multidimensional arrays.
 - **Layers**  
   Set of Mathematical operations on data.  
   Built by atoms: Dense, Convolution, Attention, Activations, etc...
 - **Loss Functions**  
   Different objectives for various applications.  
   Often used in the "Heads" of the net.
 - **Automatic Differentiation Engine**  
   Being able to calculate the Gradients of the computational graph of the net.
 - **Optimizers & Schedulers**  
   Applying update rules on the weights and step size.
 - **Data Loaders**  
   Loading data from storage, unpacking, caching, augmentation phase.
 - **Dashboard** (Optional)  
   A tool to analyze multiple experiments with nets during run time and after.
 - **Model Zoo** (Optional)  
   A set of pre defined architectures and pre trained weights.

PyTorch _claim to fame_ is its natural extension to _Python_ with its _dynamic_ (Eager) mode of operation.

## Neural Network Module (`torch.nn`)

PyTorch has a dedicated module for Neural Network elements called [`torch.nn`](https://docs.pytorch.org/docs/stable/nn.html).  
The module also defines the base class [`nn.Module`](https://docs.pytorch.org/docs/stable/generated/torch.nn.Module.html) for a definition of a Neural Network / Atom.  

The stricture of such module is:

```python
class MyModule(nn.Module):
    def __init__( self, someParam: Any, anotherParams: Any ) -> None:
        
        super().__init__() #<! Inherits from the base module
        self.oSomeElement = <Definition of an element>
        self.oAnotherElement = <Definition of an element>

    def forward( self, tX: Tensor ) -> Tensor:
        
        <Definition of the computational graph>
        
        return tY
```

The module objective is to apply a computational graph on its input.  
The way to define the computational graph requires 2 methods:

 - The `__init__()` Method  
   Defines the atoms of the _Computational Graph_.
 - The `forward()` Method  
   Defines the layout and flow of the _Computational Graph_ using the atoms.

* <font color='brown'>(**#**)</font> The `nn` module has a submodule called [`torch.nn.functional`](https://docs.pytorch.org/docs/stable/nn.functional.html).  
  It offers a stateless implementation of most of the atoms in `torch.nn`.
* <font color='brown'>(**#**)</font> The main class defines that the `__call__()` method runs the `forward()` method.  
  Hence `MyModule(tX)` will apply the `forward()` method on `tX`.

### Basic Model

$$\hat{\boldsymbol{y}}=f\left(\boldsymbol{x}\right)=\boldsymbol{W}_{3}\left(\sigma_{1}\left(\boldsymbol{W}_{1}\boldsymbol{x}\right)+\sigma_{2}\left(\boldsymbol{W}_{2}\boldsymbol{x}\right)\right)$$

<center><img src="https://media.githubusercontent.com/media/FixelAlgorithmsTeam/FixelCourses/refs/heads/master/DeepLearningMethods/2023_02/05_PyTorch/ParallelNetwork.png" style="width: 500px;"/></center>

````python
# Custom Layer
# One can define any Layer / Atom as a Module

class CustomLayer(nn.Module):
    def __init__( self, dIn: int, dOut: int ) -> None:
        super().__init__() #<! Do this to get all initialization of Layer
        
        self.oW1 = nn.Linear(dIn, dOut, bias = False)
        self.oW2 = nn.Linear(dIn, dOut, bias = False)

    def forward( self, vX: Tensor ) -> Tensor:
        
        vZ1 = torch.relu(self.oW1(vX)) #<! σ1(W1 * x)
        vZ2 = torch.relu(self.oW2(vX)) #<! σ2(W2 * x)

        return vZ1 + vZ2
````

````python
# The Model by Sequential Model

# Sequential Model
dimInW1 = 100
dimInW3 = 50
dimOut  = 10

oModel = nn.Sequential(
    CustomLayer(dimInW1, dimInW3),           #<! z = σ1(W1 * x) + σ2(W2 * x)
    nn.Linear(dimInW3, dimOut, bias = False) #<! y = W3 * z
)

torchinfo.summary(oModel, (16, 100), device = 'cpu')
````

````python
# The Model Computational Graph
# The `torchvista` package can be used to trace and display the computational graph of the model.

torchvista.trace_model(oModel, torch.randn(dimInW1))
````

````python
# Run Module

vX = torch.randn(dimInW1)
vY = oModel(vX) #<! Equivalent of `oModel.forward(vX)`

print(f'The input shape: {vX.shape}')
print(f'The output shape: {vY.shape}')
````
