### Regression:

### Simple linear regression (on the training set):

```python
from sklearn.linear_model import LinearRegression

dTrainX = dData.drop(columns='Charges')
# dTrainX = dData.drop(columns=['Charges', 'Region_northwest', 'Region_southeast', 'Region_southwest'])
dTrainY = dData['Charges']

oLinearRegresor = LinearRegression(normalize=False).fit(dTrainX, dTrainY)

vHatY   = oLinearRegresor.predict(dTrainX)
R2score = oLinearRegresor.score(dTrainX, dTrainY)

_, ax = plt.subplots(1, 2, figsize=(18, 6))
lColor = ['Smoker_yes', 'BMI']
for ii in range(2):
    ax[ii].scatter(dTrainY, vHatY, c=dTrainX[lColor[ii]])
    ax[ii].axis('tight')
    ax[ii].grid('True')
    ax[ii].set_xlabel('Charges (True)')
    ax[ii].set_ylabel('$\hat{y}$ (Predict)')
    vAxis = ax[ii].axis()
    ax[ii].plot(vAxis[:2], vAxis[:2], '--k')
    ax[ii].set_title('$R^2$ score = ' + str(R2score))

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

def PlotFit(P):
    mTrainPhi = PolynomialFeatures(degree=P, include_bias=False).fit_transform(dTrainX)

    oLinearRegresor = LinearRegression(normalize=True).fit(mTrainPhi, dTrainY)
    
    vHatY   = oLinearRegresor.predict(mTrainPhi)
    R2score = oLinearRegresor.score(mTrainPhi, dTrainY)

    _, ax = plt.subplots(1, 2, figsize=(18, 6))
    lColor = ['Smoker_yes', 'BMI']
    for ii in range(2):
        ax[ii].scatter(dTrainY, vHatY, c=dTrainX[lColor[ii]])
        ax[ii].axis('tight')
        ax[ii].grid('True')
        ax[ii].set_xlabel('Charges (True)')
        ax[ii].set_ylabel('$\hat{y}$ (Predict)')
        vAxis = ax[ii].axis()
        ax[ii].plot(vAxis[:2], vAxis[:2], '--k')
        ax[ii].set_title('$R^2$ score = ' + str(R2score))

    plt.tight_layout()
```

```python
from ipywidgets import interact, IntSlider, Layout

pSlider = IntSlider(min=1, max=10, step=1, value=1, layout=Layout(width='80%'))
interact(PlotFit, P=pSlider)
plt.show()
```

```python
mX = dTrainX.values
vY = dTrainY.values

mX = mX - np.mean(mX, axis=0)
mX = mX / np.std(mX,  axis=0)

mX.shape, vY.shape
```

```python
from sklearn.model_selection import cross_val_score, KFold

#-- Cross validation regression with Lasso:
dRes   = pd.DataFrame(columns=['P', 'R2'])
for P in [1, 2, 3]:
    mPhi                = PolynomialFeatures(P, include_bias=False).fit_transform(mX)
    mPhi                = mPhi - np.mean(mPhi, axis=0)
    mPhi[:,1:]          = mPhi[:,1:] / np.std(mPhi[:,1:], axis=0)
    vScores             = cross_val_score(LinearRegression(), mPhi, vY, cv=KFold(20, shuffle=True))
    dRes.loc[len(dRes)] = [P, np.mean(vScores)]

dRes.sort_values(by='R2', ascending=False)
```

```python
from sklearn.linear_model import Lasso

#-- Cross validation regression with Lasso:
vLam = np.logspace(-1, 1, 15)
dRes = pd.DataFrame(columns=['P', 'lam', 'R2'])
for P in [1, 2, 3]:
    for lam in vLam:
        mPhi                = PolynomialFeatures(P, include_bias=False).fit_transform(mX)
        vScores             = cross_val_score(Lasso(alpha=lam, max_iter=10000, tol=0.05), mPhi, vY, cv=KFold(20, shuffle=True))
        dRes.loc[len(dRes)] = [P, lam, np.mean(vScores)]

dRes.sort_values(by='R2', ascending=False)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

scores = cross_val_score(GradientBoostingRegressor(n_estimators=100, learning_rate=0.05), dTrainX, dTrainY, cv=KFold(20, shuffle=True))
print('GBR = ' + ': ' + str(scores.mean()))
```
